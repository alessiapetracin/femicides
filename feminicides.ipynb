In order to perform a chi-square test, we compute the expected number of feminicides per immigration from macro-region, and compare them to the actual number of feminicides per macro-region.

```r
# Z-scores:
total.feminicides <- 157 + 179 + 152 + 141 + 145; total.feminicides # 2012, 2013, 2014, 2015, 2016
```

```r
# Italians: 
e.perc.it <- 1 - 0.083 # expected percentage
e.it <- e.perc.it*total.feminicides # expected count of feminicides for italian population
cat("Expected count of feminicides for Italian population", e.it, "\n")
actual.it <- 0.745*total.feminicides # actual count of feminicides for italian pop
cat("Actual count of feminicides for Italian population", actual.it, "\n")
z.it <- (actual.it - e.it)/sqrt(e.it) # z-score for italian
cat('z-score for Italians:', z.it)
```

We can already see that there is a large absolute z-score, which indicates that the actual number of feminicides commited by Italians deviate from the number we would expect, considering the percentage of Italians on the overall population.

Since the expected count for each class should be at least 5, we cannot consider the classes 'Other' and 'South-Central Americans' in the chi-square test. We consider the other classes: Italians, Asians, North Africans and Eastern Europeans.

Through a chi-square test, we aim to find out whether the differences are strong enough to provide evidence that the author of feminicides are not a random sample of the general population- H0: the authors fo feminicides are a random sample of the general population
- H1: the authors of feminicides are not a random sample of the general population

```r
# X2 value
x2 <- z.it^2 + z.eu^2 + z.africa^2 + z.asia^2
cat('X2 value:', x2, '\n')
```

The huge chi-square value already suggests that there is evidence to support the null hypothesis is scarce.
In running the chi-square test, we consider 3 degrees of freedom, since we analyze 4 groups.

```r
# chi-squared test
1 - pchisq(x2, df = 3)
```

There is overwhelming evidence against the null hypothesis. We thus reject it.

We conduct a second analysis, testing for difference of two proportions: Italians vs non-Italians.

- H0: ethnicity of people who commit feminicides is a random sample from the general population
- H1: ethnicity of people who commit feminicides is not a random sample from the general population

```r
original_string <- immigration$Totale[197] # total foreigners
string_without_dots <- gsub("\\.", "", original_string)
foreigners.tot <- as.numeric(string_without_dots) # total population of foreigners
italian.tot <- 60665551 - foreigners.tot # total population of italians

actual.foreigners <- total.feminicides - actual.it
cat('Total feminicides committed by non-Italians:', actual.foreigners, '\n')


p.pooled <- total.feminicides/60665551 # estimate of the feminicide rate across the whole population
cat('p-pooled:', p.pooled, '\n')
# best estimate of proportions in case the null hypothesis is true

# point estimate of the difference in feminicide rate in the two groups (Italians, foreigners)
p.estimate <- actual.foreigners/foreigners.tot - actual.it/italian.tot # point estimate of the difference in femminicide rate
cat('Point estimate of the difference in feminicide rate across the two groups', p.estimate, '\n')
cat('Ratio of actual feminicides committed by foreigners and foreigner population', actual.foreigners/foreigners.tot, '\n')
cat('Ratio of actual feminicides committed by Italians and Italian population', actual.it/italian.tot, '\n')

# standard error
st.err <- sqrt(p.pooled*(1 - p.pooled)/foreigners.tot + p.pooled*(1-p.pooled)/italian.tot)
cat('Standard error:', st.err, '\n')

# confidence interval
cat('Lower bound:', p.estimate - 1.96*st.err, '\n') # lower bound
cat('Upper bound:', p.estimate + 1.96*st.err, '\n') # upper bound
# we can already see from the confidence interval that we reject the null hypothesis


# compute p-value
cat('p-value:', (1 - pnorm(p.estimate, 0, st.err))*2, '\n')
# strong evidence against the null-hypothesis
```

Both tests reject the null hypothesis: it is very unlikely that the number of feminicides by ethnicity of the murderer reflects a random sample from the general population. In fact, the probability is so low that we can approximate it to 0. Looking at the z-scores in the chi-square test, we indeed see that the actual statistics deviate quite consistently from the expected counts by ethnicity.
We can conclude that, in the general population, foreigners tend to commit more feminicides relative to the number of individuals than Italians.
In fact, foreigners are twice to three times more likely than Italians to commit a feminicide.
